```r
#Data-analytiikan ohjelmointikielet
#3. Datan muokkaus ja siivoamine
#R-esimerkit


#Ladataan tilastokeskuksen data. Haetaan data JSON-muodossa suoraan Tilastokeskuksen PXWeb-palvelun tarjoaman
#rajapinnan kautta. Asennetaan tätä varten pxweb-paketti. 

#install.packages("pxweb")

library(pxweb)

url = "https://pxdata.stat.fi:443/PxWeb/api/v1/fi/StatFin/aku/statfin_aku_pxt_12dz.px"


data <- pxweb_get(url)

#Tulostetaan sisältö

str(data)
```

```r
#Tulostuksesta nähdään käytetyt koodit. Tehdään kysely, jossa nämä kaikki ovat mukana. 
#Tämä kysely kohdistetaan sitten JSON-tiedostoon. Haetaan tiedostosta siis kaikki mahdollinen. 

query <- list(
  "Vuosi" = "*",
  "Ikä" = "*",
  "Tiedot" = "*"
)

data <- pxweb_get_data(url, query = query)

print(data)
```

```r
#Muunnetaan data tibbleksi (kehittyneempi dataframe, vastaa Julian dataframea)

df <- as.data.frame(data, column.name.type = "text", variable.value.type = "text")

df
```

```r
#Nyt huomataan, että pxweb-kirjasto ymmärtää rakenteen semantiikan, kun taas Julia lukee vain raakadatan. Tämän takia arvot ovat oikein
#ja Juliassa tehtyä muokkausta ei tarvita. Näin ollen R on monesti parempi valinta PxWebin tarkasteluun.

#Lisäksi huomataan, että tuloksissa on mukana Yhteensä-arvoja. Suodatetaan nämä pois. Lisäksi

library(dplyr)

df <- df |> filter(Ikä != "Yhteensä")

df
```

```r
#Nyt huomataan, että dataframe on oletuksena lyhyessä (leveässä) muodossa toisin kuin Juliassa. Muutetaan se pitkäksi. 

library(tidyr)

df_pitka <- pivot_longer(
  df,
  cols = -c(Vuosi, Ikä),   # kaikki muut sarakkeet paitsi nämä venytetään
  names_to = "Tiedot",     # uusi sarake sarakenimistä
  values_to = "Arvo"       # uusi sarake arvoista
)

df_pitka


#Funktio pivot_wider() muuttaa pitkän datan tarvittaessa lyhyeksi. 
```

```r
#Analysoidaan dataa seuraavaksi tekemällä Pivot-taulukko, jossa osa datasta on suodatettu pois. 

df_suod <- df_pitka %>%
  filter(
    Vuosi == "2022",
    Tiedot %in% c(
      "Aikuiskoulutukseen osallistuneet, %",
      "Työhön tai ammattiin liittyvään aikuiskoulutukseen osallistuneet, %",
      "Aikuiskoulutukseen osallistuneet, lkm"
    )
  ) %>%

select(Vuosi, Ikä, Tiedot, Arvo) %>%
  pivot_wider(
    names_from = Tiedot,
    values_from = Arvo
  )
  
#Tehtävä: Käytä tekoälyä ja selvitä yksityiskohtaisesti mitä yllä olevassa koodissa tapahtuu. Tämän ymmärtäminen on tärkeää. 
df_suod
```

```r
#Lasketaan lopuksi montako henkilöä osallistui aikuiskoulutukseen vuonna 2022. Laskeminen olisi onnistunut myös
#ilman tyyppimuunnosta, koska kaikki sarakkeen arvot ovat kokonaislukuja. Jos näin ei olisi, niin laskeminen antaisi
#virheilmoituksen. Myös desimaaliluvut kelpaisivat. 

sum(df_suod$`Aikuiskoulutukseen osallistuneet, lkm`, na.rm = TRUE) #Argumentti na.rm = TRUE varmistaa, että puuttuvia arvoja ei oteta mukaan. 



```
