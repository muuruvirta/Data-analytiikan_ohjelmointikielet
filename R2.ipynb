```r
#Data-analytiikan ohjelmointikielet
#2. Datan hakeminen ja rakenteiden hallinta
#R-esimerkit

#Luetaan CSV-tiedosta ja tulostetaan se tarkasteltavaksi. Otetaan käyttöön kirjasto readr, jotta saadaan "moderni" dataframe eli tibble käyttöön.
#Kannattaa asentaa tidyverse-paketti, johon myös readr sisältyy. Tidyverse sisältää kirjastoja, joita käytämme myöhemmin opintojaksolla. 
#Koska tiedosto on UTF8-koodattu, käytetään funktiota read_csv2.

library(readr)
data <- read_csv2("C:/Users/Installer/Desktop/data.csv")
data
```

```r
#Huomataan, että tiedoston lukeminen halutulla tavalla ei onnistunut. Funktio read_csv2 olettaa, että erottiminena on puolipiste ja 
#käytetään desimaalipilkkua. Käytetään sen sijaan funktiota read_delim(), johon voidaan antaa tiedoston koodaus ja muut tiedot. 

data <- read_delim(
  "C:/Users/Installer/Desktop/data.csv", delim = ";",  # tietoerotin on puolipiste
  locale = locale(encoding = "UTF-8", decimal_mark = "."),  # UTF-8 ja piste desimaalina
  skip = 2  # Otsikko on kolmannella rivillä
)
  
data
```

```r
#Huomataan, että sarakeotsikoiden nimissä on puutteita. Luetaan data vielä kerran sisään ja estetään R:ää muokkaamasta otsikoita.

data <- read.csv( "C:/Users/Installer/Desktop/data.csv", sep = ";", #erotinmerkki on ;
    fileEncoding = "UTF-8", #koodaus on UTF-8
    skip = 2, check.names = FALSE) #skipataan kaksi ensimmäistä riviä ja tulkitaan kolmas rivi otsikoksi
  
data

```

```r
#Poistetaan seuraavaksi ne sarakkeet, joiden kaikissa soluissa on pelkkä piste (.). Nämä soluthan ovat käytännössä puuttuvia arvoja. 
#Poistetaan lisäksi 1. sarake, joka on tarpeeton. 

#sapply(data,....) käy kaikki sarakkeet läpi
#all(col=".") palauttaa

data <- data[, !sapply(data, function(col) all(col == "."))]

data <- data[,-1] #Otetaan mukaan kaikki rivit, mutta poistetaan 1. sarake.

data
```

```r
#Talletetaan seuraavaksi siivottu tiedosto kovalevylle tulevaa käyttöä varten. Funktio write_csv2 käyttää erottimimena
#puolipistettä ja ei muuta desimaalierotinta.

write_csv(
  data,
  "C:/Users/Installer/Desktop/siivottudataR.csv"
)
```

```r
#Tarkastellaan lopuksi dataa vähän. Katsotaan ensin onko puuttuvia arvoja (NA).

anyNA(data)

#Puuttuvien arvojen yhteismäärä

sum(is.na(data))

#Puuttuvien arvojen määrä jokaisessa sarakkeessa

colSums(is.na(data))

#Puuttuvien arvojen määrä riveittäin

rowSums(is.na(data))
```

```r
#Kaksi viimeistä tulosta yllä ovat aika vaikeita tulkita. Muutetaan colSums-funktion antama tulos tibbleksi. Tibble-sisältyy 
#aiemmin asennettuun tidyverse-pakettiin.


library(tibble)

sarakkeiden_puuttuvat_arvot <- enframe(colSums(is.na(data)),
                           name = "Sarake",
                           value = "Puuttuvia")

print(sarakkeiden_puuttuvat_arvot)
```
