```julia
#Data-analytiikan ohjelmointikielet
#2. Datan hakeminen ja rakenteiden hallinta
#Julia-esimerkit

#Otetaan käyttöön CSV-kirjasto, jota käytetään CSV-tiedostojen käsittelyyn. Mikäli data olisi Excel-tiedostossa
#oikea kirjasto olisi XLSX. Otetaan lisäksi käyttöön DataFrames-kirjasto, joka sisältää tarvittavat asiat datan käsittelyyn.
#Muista ensin asentaa tarvittavat paketit Juliaan. 

using CSV, DataFrames


#Luetaan Excel-tiedosto ja muunnetaan luettu data DataFrame-tyyppiseksi. (Muuta tiedostopolku vastaamaan omaasi.) Tässä oletetaan, että csv-tiedosto
#on tallennettu utf8-muodossa (skandit näkyvät oikein). Jos alkuperäisen tiedoston koodaus on jotain muuta, niin muunna se utf8-muotoon. Helpoiten tämä
#onnistuu avaamalla tiedosto Muistiolla ja tallentamalla se oikeaan muotoon. Muuntaminen on mahdollista myös Juliassa. 
#ÄLÄ avaa tiedostoa Excelissä. Excel muuntaa desimaalilukuja automaattisesti päivämääriksi. 

#Data on peräisin Tilastokeskukselta.

#data = XLSX.readtable("tiedot.xlsx", "Data") |> DataFrame (Mikäli data olisi Excel-muodossa.)

data = CSV.read("C:\\Users\\Installer\\Desktop\\data.csv",DataFrame, delim = ';') 
#delim-argumentilla varmistetaan, että Julia tunnistaa erotinmerkin oikein. 
#Varoituksesta ei tarvitse tässä välittää. 
```

```julia
#Nyt huomataan, että osa sarakkeiden tietotyypeistä on mitä sattuu ja lisäksi osa desimaaliluvuista esitetään päivämääränä. 
#Haetaan data uudelleen siten, että pakotetaan sarakkaiden tietotyypit. Koska csv-tiedostoa luettaessa tehdään automaattista
#tietotyyppien tunnistusta, tässä tapauksessa yhden sarakkeen määrittäminen riittää. Tarkista aina tietotyyppien muunnoksen
#jälkeen, että kaikissa sarakkeissa on oikea tietotyyppi. 

#Laitetaan samalla rivillä 3 olevat sarakeotsikot oikealle paikalle header-argumentilla. Tarkista datatiedostosta, millä rivillä otsikkodata sijaitsee.
#Dict-tietotyyppi on sanakirja.

data = CSV.read("C:\\Users\\Installer\\Desktop\\data.csv",DataFrame, header = 3, types = Dict("Väkiluvun muutos edellisestä vuodesta, %" => Float64)) 
```

```julia
#Poistetaan seuraavaksi 1. sarake, koska siinä on vain vuosiluku. 

select!(data, Not("Vuosi"))

```

```julia
#Poistetaan seuraavaksi kaikki ne sarakkeet, joissa ei ole yhtään arvoa. Juliassa on helpohkoa poistaa pelkkiä puuttuvia
#arvoja (missing) sisältävä  rivi / sarake. Tilastokeskuksen portaali laittaa kuitenkin puuttuvan arvon tilalle pisteen (.)
#tiedostoa ladattaessa. Tämä on hyvä tiedostaa. 

# Apufunktio, joka tutkii milloin solussa ei ole arvoa. Tämä funktio ottaa huomioon sekä puuttuvan arvon että pisteen. 

eiarvoa(x) = ismissing(x) || x == "." || (x isa AbstractString && isempty(strip(x)))

# Etsitään sarakkeet, joissa kaikki arvot ovat tyhjiä/pisteitä/missing. Juliassa hakasuluissa esitellään vektori. 

tyhja_sarake = [c for c in names(data) if all(x -> eiarvoa(x), data[!, c])]

# Poista ne sarakkeet pysyvästi
select!(data, Not(tyhja_sarake))


```

```julia
#Nyt data on saatu siivottua ja se on valmis analysoitavaksi. Talletetaan se tulevaisuuden käyttöä varten uuteen csv-tiedostoon. 
#Julia käyttää oletuksena UTF8-koodausta.

CSV.write("C:\\Users\\Installer\\Desktop\\siivottu_data.csv", data; delim=';')


```

```julia
#Tarkastellaan vielä lopuksi siivottua dataa vähän. 

#Tutkitaan onko datassa puuttuvia arvoja. Seuraava komento tutkii koko dataframen ja palauttaa True, jos niitä löytyy. 

any(ismissing, eachcol(data))
```

```julia
#Siivottuja arvoja ei löytynyt, mutta tutkitaan esimerkin takia kuitenkin vielä sarake kerrallaan. Seuraava komento palauttaa
#vektorin, jossa puuttuvien arvojen lukumäärä samassa järjestyksessä kuin names() -komento antaa. Muistathan, että hakasulut
#tarkoittavat Juliassa vektoria. 

print(names(data))
[sum(ismissing, col) for col in eachcol(data)]
```

```julia
#Sarake-otsikot ja puuttuvien arvojen lukumäärä voidaan yhdistää käyttämällä sanakirja-tietotyyppiä (Dict).
Dict(name => sum(ismissing, data[!, name]) for name in names(data))


```
